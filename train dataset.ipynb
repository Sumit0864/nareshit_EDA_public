### outlier analysis

```python
q1=np.percentile(train_df['ApplicantIncome'],25)
q2=np.percentile(train_df['ApplicantIncome'],50)
q3=np.percentile(train_df['ApplicantIncome'],75)
q1,q2,q3
```

```python
IQR=q3-q1
IQR
```

```python
lb=q1-1.5*IQR
ub=q3+1.5*IQR
lb,ub
```

```python
np.mean(train_df['ApplicantIncome'])
```

```python
c1=train_df['ApplicantIncome']<lb
c2=train_df['ApplicantIncome']>ub
con=c1|c2
outlier_df=train_df[con]
outlier_df
```

```python
c1=train_df['ApplicantIncome']>lb
c2=train_df['ApplicantIncome']<ub
con=c1&c2
non_outlier_df=train_df[con]
non_outlier_df
```

```python
train_df['ApplicantIncome'].size,non_outlier_df['ApplicantIncome'].size
```

### dealing the outliers

```python
#drop
#impute median values
# cap with q1and q3 values
```

```python
#impute median manually
l1=[]
median=np.median(train_df['ApplicantIncome'])
for i in train_df['ApplicantIncome']:
    if i<lb or i>ub:
        l1.append(median)
    else:
        l1.append(i)
train_df['ApplicantIncome_new']=l1
```

```python
train_df
```

```python
plt.subplot(2,2,1).hist(train_df['ApplicantIncome'],bins=40)
plt.subplot(2,2,2).hist(train_df['ApplicantIncome_new'],bins=40)
plt.show()
```

```python
plt.hist(train_df['ApplicantIncome'],bins=40)
plt.hist(non_outlier_df['ApplicantIncome'],bins=40)
plt.show()
```

```python
# using np.where()
c1=train_df['ApplicantIncome']<lb
c2=train_df['ApplicantIncome']>ub
con=c1|c2
trw=median
fls=train_df['ApplicantIncome']
train_df['ApplicantIncome']=np.where(con,trw,fls)
```
